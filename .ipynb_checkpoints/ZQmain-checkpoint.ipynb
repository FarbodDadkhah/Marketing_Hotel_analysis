## Project-Background-and-Objective-of-Analysis  

The primary goal of this analysis is to assess guest behavior, with a focus on nationality and party type, to provide data-driven recommendations that will inform targeted advertising campaigns for hotels. By understanding the characteristics of guests from different nationalities, we aim to enhance customer acquisition strategies and optimize profitability.

The analysis will focus on a target audience drawn from three key countries: the United States, the United Kingdom, and Italy, as specified. However, the sample may be adjusted to focus on a subset of these countries, depending on further strategic conclusions.

We will design a strategic framework for efficiently allocating the advertising budget to maximize return on investment, ensuring that resources are directed toward the most promising customer segments for growth and profitability. 
the framework shall be used by a hotel growth consulting company to give both hotels, useful guidance.


## Executive Summary

### **Recommendations to The Executive & Marketing team:**

- For both hotels, strategic marketing and promotional efforts should **prioritize the UK market** to capitalize on the existing strong customer base and enhance engagement further.

- **Both hotels have experienced a consistent growth in family guest bookings**. Given that family guests generate a higher rate of room revenue per booking, it is recommended to target this segment more effectively through the hotels' advertising campaigns.

- **Focus on Retaining Offline TA Partnerships:**
   - Given their significant contribution (56%) and near-zero cancellation rate, prioritize maintaining strong relationships with offline travel agencies.

- **Improve Online TA Booking Experience:**
   - Address the higher cancellation rate (32%) for online bookings by analyzing user feedback, streamlining the booking process, and offering flexible options to increase conversion rates.

- **Target Seasonal Campaigns:**
   - For UK families, emphasize marketing efforts during **May to August** to capitalize on peak booking periods.
   - For UK adults, consider promotions or incentives during **January**, **November**, and **December** to boost attendance during less popular months.

- **Expand Distribution Channels:**
   - While focusing on high-performing channels, explore opportunities to diversify distribution strategies, such as partnerships with niche travel platforms or direct marketing to repeat customers.


**Recommendations to the Data Team:**

- It is strongly recommended to enhance **data collection** and **recording of customer spending** within the hotel. The rationale for this is as follows:
   - By obtaining detailed spending data from guests, we could gain a deeper understanding of the profitability of **UK guests**. This insight would allow us to offer more tailored services, which is an exceptionally efficient strategy for our hotels, as **service margins** in the hospitality industry tend to be high globally. Leveraging this data could significantly boost profits by delivering targeted services based on customer behavior.

```python
# Helper functions
def feature_engineering_main_tables(df):
     # 
    df['total_stay'] = df.loc[:,'stays_in_week_nights'] + df.loc[:,"stays_in_weekend_nights"]

    df['kids'] = df['children'] + df['babies'] # -> lets wrap up the number of minors as "kids"
    
    df['party_size'] = df['adults'] + df['kids'] # ->  useful feature is total number of guests
    
     # loyalty cat
    df['loyalty_tier'] = pd.cut(
    df['previous_bookings_not_canceled'],
    bins=[-1, 0, 3, 10, float('inf')],  # -> ranges for each tier:
    labels=['New', 'Occasional', 'Frequent', 'VIP'] )
    
    # total room rev
    df['room_rev'] = df['total_stay'] * df['adr'] * df['party_size']
    return df
def write_updated_dataset(df,PATH):
    df.to_csv(f'./data/{PATH}.csv', index=False)
```

```python
''' 
Fix Data Format, Type and logic + apply Feature engineer function.
'''
City_booking_data = feature_engineering_main_tables(City_booking_data)
City_booking_data = City_booking_data[ (City_booking_data['adr']>30) & (City_booking_data['total_stay']>0)]

City_booking_data = City_booking_data.reset_index() 
City_booking_data["id"] = City_booking_data.index


RESORT_booking_data = feature_engineering_main_tables(RESORT_booking_data)
RESORT_booking_data = RESORT_booking_data[ (RESORT_booking_data['adr']>30) & (RESORT_booking_data['total_stay']>0)]

RESORT_booking_data = RESORT_booking_data.reset_index() 
RESORT_booking_data["id"] = RESORT_booking_data.index


```

### revenue generation for uk customers per booking

**Key Insights:**

1. The value of **UK customers** is significantly high and well-justified for both **City Hotels** and **Resort Hotels**.
2. On average, **UK customers** generate approximately 43% more revenue from room bookings in **Resort Hotels** compared to other nationalities.

```python
grouped__City__avg__totaly_stay = City_booking_data.groupby('country')['total_stay'].mean()
grouped__City__avg__party_size = City_booking_data.groupby('country')['party_size'].mean()
grouped__City__avg__adr = City_booking_data['adr'].mean()

total_avg_revenue_room = round(
    (grouped__City__avg__totaly_stay * grouped__City__avg__party_size * grouped__City__avg__adr)
    ,1)

UK_average_revenue_room = total_avg_revenue_room['GBR']
print('City hotel:')
print(f'Overall average revenue: ${round(total_avg_revenue_room.mean())} vs UK average revenue: ${UK_average_revenue_room}')

#Resort
RESORT_booking_data = feature_engineering_main_tables(RESORT_booking_data)

grouped__Resort__avg__totaly_stay = RESORT_booking_data.groupby('country')['total_stay'].mean()
grouped__Resort__avg__party_size = RESORT_booking_data.groupby('country')['party_size'].mean()
grouped__Resort__avg__adr = RESORT_booking_data['adr'].mean()

total_avg_revenue_room_RESORT = round(
    (grouped__Resort__avg__totaly_stay * grouped__Resort__avg__party_size * grouped__Resort__avg__adr)
    ,1)

UK_average_revenue_room_RESORT = total_avg_revenue_room_RESORT['GBR']
print('Resort hotel:')
print(f'Overall average revenue: ${round(total_avg_revenue_room_RESORT.mean())} vs UK average revenue: ${UK_average_revenue_room_RESORT}')

# Bar Plot: UK vs Overall Average Revenue
categories = ['City Hotel', 'Resort Hotel']
uk_avg = [UK_average_revenue_room, UK_average_revenue_room_RESORT]
overall_avg = [total_avg_revenue_room.mean(), total_avg_revenue_room_RESORT.mean()]

# plot initiate
fig, ax = plt.subplots(figsize=(16, 6))
x = range(len(categories))
ax.bar(x, overall_avg, width=0.4, label='Overall Average Revenue', color='skyblue')
ax.bar([p + 0.4 for p in x], uk_avg, width=0.4, label='UK Average Revenue', color='salmon')

# plot format
ax.set_xticks([p + 0.2 for p in x])
ax.set_xticklabels(categories)
ax.set_ylabel('Average Revenue ($)')
ax.set_title('UK vs Overall Average Revenue')
ax.legend()
plt.show()
```

### booking lead time seasonal attendance trends and distribution partner performance

#### **Key Insights**

1. **Average Lead Time:**
   - The average lead time for bookings among both family and adult-only guests is approximately **110 days**.

2. **UK Families' Booking Patterns:**
   - UK families predominantly book their stays during the months of **May to August**, indicating a strong seasonal preference.

3. **UK Adults' Attendance Trends:**
   - UK adult guests exhibit a balanced distribution of attendance throughout the year, with the exception of **January**, **November**, and **December**, which appear to be less appealing for this group.

4. **Distribution Channels Performance:**
   - **Online Travel Agencies (TA):** Contribute approximately **27%** of the bookings, with a cancellation rate of **32%**, which is significantly higher than the average.
   - **Offline Travel Agencies (TA):** Account for around **56%** of the bookings, with an exceptionally low cancellation rate nearing **0%**.
   - Combined, these channels contribute to **83%** of the total intake across both hotels, demonstrating their dominance over other distribution and marketing avenues.


```python
# OUTER MERGE, successfull bookings(not canceled)
UK_RESORT  = RESORT_booking_data[(RESORT_booking_data['country'] =='GBR') & (RESORT_booking_data['is_canceled'] == 0)]
UK_CITY = City_booking_data[(City_booking_data['country'] =='GBR') & (City_booking_data['is_canceled'] == 0)]
data_merged = pd.merge(UK_RESORT,UK_CITY,how='outer')

booked_per_months = data_merged.groupby('arrival_date_month')["id"].count().sort_values(ascending=False)
booked_per_months_family = data_merged[data_merged['kids'] > 0].groupby('arrival_date_month')["id"].count().sort_values(ascending=False)
booked_per_months_adults = data_merged[data_merged['kids'] == 0].groupby('arrival_date_month')["id"].count().sort_values(ascending=False)


############################################################
family_avg_leadtime = data_merged[data_merged['kids'] > 0 ]['lead_time'].mean()
adult_avg_leadtime = data_merged[data_merged['kids'] == 0 ]['lead_time'].mean()

print(f'UK Adult guests average leadtime: {round(adult_avg_leadtime,0)}')
print(f'UK Family guests average leadtime: {round(family_avg_leadtime,0)}')

```

### using pearson correlation analysis to find potential linear relationships between features

    Findings after analysis : there is no meaningfull correlation between our features/ engineered features.

```python
import warnings
# for non-neccesary pandas next versions warnings
warnings.filterwarnings('ignore')

# copy so we only apply further data preprocessing on the copy
data = data_merged.copy()
data = feature_engineering_main_tables(data)
# Fill missing values
for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        data[column].fillna(data[column].median(), inplace=True)

#  Standardize numeric features
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = (data[numeric_columns] - data[numeric_columns].mean()) / (data[numeric_columns].std() + 1e-10)

# 
numeric_data = data.select_dtypes(include=[np.number])
numeric_data = numeric_data.drop(columns=['is_canceled','previous_cancellations'])
correlation_matrix = numeric_data.corr(method='pearson')

num_features = len(numeric_data.columns)
heatmap_size = (min(20, num_features), min(20, num_features)) 
plt.figure(figsize=heatmap_size)
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='coolwarm', 
)
plt.show()

```

-- Use a CTE to calculate total revenue by nationality and filter by a threshold \
WITH revenue_by_nationality AS ( \
    SELECT g.nationality,  \ 
           SUM(b.revenue) AS total_revenue \
    FROM bookings AS b \
    JOIN guests AS g ON b.guest_id = g.id \
    GROUP BY g.nationality \
) 



SELECT nationality, total_revenue \
FROM revenue_by_nationality \

  AND nationality IN ('GBR', 'USA', 'ITA') \
ORDER BY total_revenue DESC;

-- Rank guests by their total revenue within each nationality using a window function \
SELECT g.nationality,  \
       g.id,  \
       SUM(b.revenue) AS total_revenue, \
       RANK() OVER (PARTITION BY g.nationality ORDER BY SUM(b.revenue) DESC) AS rank_within_nationality \
FROM bookings AS b \
JOIN guests AS g ON b.guest_id = g.id \
GROUP BY g.nationality, g.id \
ORDER BY g.nationality, rank_within_nationality;
